Test using Answer Relevancy Metric

```python
import os

import deepeval
from dotenv import load_dotenv, find_dotenv
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.models import DeepEvalBaseLLM
from groq import Groq
from groq.types.chat import ChatCompletionUserMessageParam

# Auto finds the .env file (which should be at root level of project)
load_dotenv(find_dotenv())
print("Key loaded:", os.getenv("GROQ_API_KEY") is not None) # Should return True

class GroqModel(DeepEvalBaseLLM):
    def __init__(self):
        self.client = Groq(api_key=os.getenv("GROQ_API_KEY"))
        super().__init__()

    def get_model_name(self):
        return "groq/compound"

    def load_model(self):
        return self.client

    def generate(self, prompt: str, schema=None):
        messages: list[ChatCompletionUserMessageParam] = [
            ChatCompletionUserMessageParam(role="user", content=prompt)
        ]
        response = self.client.chat.completions.create(
            model="groq/compound",
            messages=messages,
            stream=False
        )
        return response.choices[0].message.content, 0

    async def a_generate(self, prompt: str, schema=None):
        return self.generate(prompt, schema)

groq_model = GroqModel()

answer_relevancy_metric = AnswerRelevancyMetric(model=groq_model)
test_case = LLMTestCase(
    input="Who is the current president of United States of America?",
    actual_output="Donald Trump",
    retrieval_context=["Donald Trump serves as the current president of America."]
)

answer_relevancy_metric.measure(test_case)
print(answer_relevancy_metric.score)
```

Evaluate our testcase using Evaluate()

```python
import os
from dotenv import load_dotenv, find_dotenv
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.models import DeepEvalBaseLLM
from groq import Groq
from groq.types.chat import ChatCompletionUserMessageParam
from deepeval.evaluate import evaluate

# Auto finds the .env file (which should be at root level of project)
load_dotenv(find_dotenv())
print("Key loaded:", os.getenv("GROQ_API_KEY") is not None) # Should return True

class GroqModel(DeepEvalBaseLLM):
    def __init__(self):
        self.client = Groq(api_key=os.getenv("GROQ_API_KEY"))
        super().__init__()

    def get_model_name(self):
        return "groq/compound"

    def load_model(self):
        return self.client

    def generate(self, prompt: str, schema=None):
        messages: list[ChatCompletionUserMessageParam] = [
            ChatCompletionUserMessageParam(role="user", content=prompt)
        ]
        response = self.client.chat.completions.create(
            model="groq/compound",
            messages=messages,
            stream=False
        )
        return response.choices[0].message.content, 0

    async def a_generate(self, prompt: str, schema=None):
        return self.generate(prompt, schema)

groq_model = GroqModel()

answer_relevancy_metric = AnswerRelevancyMetric(model=groq_model, async_mode=False)
test_case = LLMTestCase(
    input="Who is the current president of United States of America?",
    actual_output="Donald Trump",
    retrieval_context=["Donald Trump serves as the current president of America. "]
)

evaluate(test_cases=[test_case], metrics=[answer_relevancy_metric])
```
